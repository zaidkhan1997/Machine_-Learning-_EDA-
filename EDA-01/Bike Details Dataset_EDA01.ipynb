```python
#Question_No.10:What is the relationship between selling_price and km_driven for first-owner bikes.

#Answer:
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'BIKE DETAILS.csv' 
df = pd.read_csv(file_path)

# Display the first few rows to inspect the structure (optional)
print(df.head())

# Assume the column for ownership type is 'ownership_type' and the columns for selling price and km driven are 'selling_price' and 'km_driven' (adjust if needed)
# Filter the dataset for first-owner bikes
first_owner_bikes = df[df['owner'] == 'First Owner']

# Plotting the relationship between selling price and km driven for first-owner bikes
plt.figure(figsize=(10, 6))
sns.scatterplot(data=first_owner_bikes, x='km_driven', y='selling_price', color='blue')

# Adding title and labels
plt.title('Relationship Between Selling Price and Km Driven for First-Owner Bikes', fontsize=14)
plt.xlabel('Kilometers Driven', fontsize=12)
plt.ylabel('Selling Price', fontsize=12)

# Show the plot
plt.show()

# You can also calculate the correlation coefficient to quantify the relationship
correlation = first_owner_bikes[['km_driven', 'selling_price']].corr()
print("Correlation between km_driven and selling_price for first-owner bikes:")
print(correlation)
```

```python
#Question_No.11: Identify and remove outliers in the km_driven column using the IQR methodB

#Answer:
import pandas as pd

# Load the dataset
file_path = 'BIKE DETAILS.csv' 
df = pd.read_csv(file_path)

# Display the first few rows to inspect the structure (optional)
print(df.head())

# Assume the column for kilometers driven is named 'km_driven' (adjust if needed)
# Calculate the Q1 (25th percentile) and Q3 (75th percentile) for the km_driven column
Q1 = df['km_driven'].quantile(0.25)
Q3 = df['km_driven'].quantile(0.75)

# Calculate the IQR
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify the outliers
outliers = df[(df['km_driven'] < lower_bound) | (df['km_driven'] > upper_bound)]

# Remove the outliers
df_no_outliers = df[(df['km_driven'] >= lower_bound) & (df['km_driven'] <= upper_bound)]

# Output the result
print(f"Number of outliers identified: {len(outliers)}")
print(f"Data shape before removing outliers: {df.shape}")
print(f"Data shape after removing outliers: {df_no_outliers.shape}")
```

```python
#Question_No.13:What is the average depreciation in selling price based on the bike's age (current year - manufacturing
#year).

#Answer:
import pandas as pd
import datetime

# Load the dataset
file_path = 'BIKE DETAILS.csv'
df = pd.read_csv(file_path)

# Display the first few rows to inspect the structure (optional)
print(df.head())

# Get the current year
current_year = datetime.datetime.now().year

# Calculate the bike's age (current year - manufacturing year)
df['bike_age'] = current_year - df['year']

# Assuming 'ex_showroom_price' is the original selling price and 'selling_price' is the current selling price
# Calculate the depreciation (difference between original price and current price)
df['depreciation'] = df['ex_showroom_price'] - df['selling_price']

# Group by bike's age and calculate the average depreciation
average_depreciation_by_age = df.groupby('bike_age')['depreciation'].mean()

# Output the result
print("Average depreciation in selling price based on bike's age:")
print(average_depreciation_by_age)
```

```python
#Question_No.14: Which bike names are priced significantly above the average price for their manufacturing year.

#Answer:
import pandas as pd

# Load the dataset
file_path = 'BIKE DETAILS.csv'
df = pd.read_csv(file_path)

# Display the first few rows to inspect the structure (optional)
print(df.head())

# Check the correct column name for bike name (assuming it's 'name' here)
bike_name_column = 'name'  # Replace this with the correct column name if needed

# Calculate the average selling price and standard deviation for each manufacturing year
average_price_by_year = df.groupby('year')['selling_price'].mean()
std_price_by_year = df.groupby('year')['selling_price'].std()

# Set a threshold for significantly above average (e.g., 1 standard deviation above the average price)
threshold = average_price_by_year + std_price_by_year

# Merge the average price and threshold back into the original dataset
df = df.merge(average_price_by_year, on='year', suffixes=('', '_avg'))
df = df.merge(std_price_by_year, on='year', suffixes=('', '_std'))

# Identify bikes whose selling price is above the threshold
above_average_bikes = df[df['selling_price'] > df['selling_price_avg'] + df['selling_price_std']]

# Output the bike names and their corresponding prices
print("Bikes priced significantly above the average price for their manufacturing year:")
print(above_average_bikes[[bike_name_column, 'year', 'selling_price']])
```

```python
#Question_No.15: Develop a correlation matrix for numeric columns and visualize it using a heatmap.

#Answer:
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'BIKE DETAILS.csv'
df = pd.read_csv(file_path)

# Display the first few rows to inspect the structure (optional)
print(df.head())

# Select only the numeric columns
numeric_columns = df.select_dtypes(include=['number']).columns

# Calculate the correlation matrix for numeric columns
correlation_matrix = df[numeric_columns].corr()

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

# Adding title and labels
plt.title('Correlation Matrix Heatmap', fontsize=16)
plt.show()
```
